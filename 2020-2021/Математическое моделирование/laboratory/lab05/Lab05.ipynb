```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
a=0.31
b=0.32
c=0.054
d=0.055

x0=[7,15]

t=np.arange(0, 400, 0.1)
```

```python
def dx(x, t):
    dx1= -a*x[0]+c*x[0]*x[1]
    dx2= b*x[1]-d*x[0]*x[1]
    return [dx1, dx2]
```

```python
y=odeint(dx, x0,t)
y
```

```python
plt.title("Популяция хищников")
plt.plot(t, y[:, 0])
```

```python
plt.title("Популяция жертв")
plt.plot(t, y[:, 1])
```

```python
plt.title("Фазовый портрет")
plt.plot(y[:, 0], y[:, 1])
```
