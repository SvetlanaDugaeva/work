```python
import math
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as pltl
```

```python
s=11.8
v=4.2
```

```python
fi=math.pi/3
r0=s/(v+1)
```

```python
def dr(r, tetha):
    dr=r/math.sqrt(v*v-1)
    return dr
```

```python
tetha0=0
tetha=np.arange(tetha0,2*math.pi,0.01)
r=odeint(dr,r0,tetha)
```

```python
def lod(t):
    x=math.tan(fi+math.pi)*t
    return x
```

```python
def kat(t1):
    x=math.tan(tetha0)*t1
    return x
```

```python
t=np.arange(0,10,1)
t1=np.arange(r0,s,0.01)
```

```python
def pol(x,y):
    ropol=np.sqrt(x**2+y**2)
    fipol=np.arctan2(y,x)
    return ropol, fipol
```

```python
r1, tetha1=pol(t,lod(t))
r2, tetha2=pol(t1,kat(t1))
pltl.polar(tetha,r,'g')
pltl.polar(tetha2,r2,'g')
pltl.polar(tetha1,r1,'b')
```

```python
r0=s/(v-1)
tetha0=-math.pi
tetha=np.arange(tetha0,-tetha0,0.01)
r=odeint(dr,r0,tetha)
```

```python
t=np.arange(0,10,1)
t1=np.arange(-s,-r0,0.01)
```

```python
r1, tetha1=pol(t,lod(t))
r2, tetha2=pol(t1,kat(t1))
pltl.polar(tetha,r,'g')
pltl.polar(tetha2,r2,'g')
pltl.polar(tetha1,r1,'b')
```
