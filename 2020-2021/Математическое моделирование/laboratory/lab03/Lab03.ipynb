```python
import numpy as np
from math import sin, cos
from scipy.integrate import odeint
import matplotlib.pyplot as pltl
```

```python
y0=92000
x0=202000
a1=0.13
b1=0.51
c1=0.41
h1=0.15
a2=0.08
b2=0.76
c2=0.64
h2=0.07
t0=0
tmax=1
dt=0.05
t=np.arange(t0,tmax,dt)
v0=np.array([x0,y0])
```

```python
def P1(t):
    p1=(0.5*sin(t+13))
    return p1
```

```python
def Q1(t):
    q1=(0.5*cos(t+2))
    return q1
```

```python
def P2(t):
    p2=sin(2*t)+1
    return p2
```

```python
def Q2(t):
    q2=cos(3*t)+1
    return q2
```

```python
def dy1(y1,t):
    dy11=-a1*y1[0]-b1*y1[1]+P1(t)
    dy12=-c1*y1[0]-h1*y1[1]+Q1(t)
    return [dy11, dy12]
```

```python
def dy2(y2,t):
    dy21=-a2*y2[0]-b2*y2[1]+P1(t)
    dy22=-c2*y2[0]*y2[1]-h2*y2[1]+Q1(t)
    return [dy21, dy22]
```

```python
y1=odeint(dy1,v0,t)
data1=[[elem[0] for elem in y1],[elem[1] for elem in y1]]
```

```python
y2=odeint(dy2,v0,t)
data2=[[elem[0] for elem in y2],[elem[1] for elem in y2]]
```

```python
pltl.plot(t,data1[0],'blue',label='X')
pltl.plot(t,data1[1],'green',label='Y')
pltl.title('Модель боевых действий между регулярными войсками')
pltl.xlabel('Время')
pltl.ylabel('Численность армии')
pltl.ylim(0,None)
pltl.legend()
pltl.grid(True)
pltl.margins(0.05)
pltl.subplots_adjust(left=0,bottom=0,right=1, top=1)
```

```python
pltl.plot(t,data1[0],'blue',label='X')
pltl.plot(t,data2[1],'green',label='Y')
pltl.title('Модель боевых действий между регулярными войсками')
pltl.xlabel('Время')
pltl.ylabel('Численность армии')
pltl.ylim(0,None)
pltl.legend()
pltl.grid(True)
pltl.margins(0.05)
pltl.subplots_adjust(left=0,bottom=0,right=1, top=1)
```
