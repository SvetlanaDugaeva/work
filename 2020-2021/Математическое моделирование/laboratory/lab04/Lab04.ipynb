```python
import numpy as np
import matplotlib.pyplot as pltl
from math import sin, cos
from scipy.integrate import odeint
```

Колебания гармонического осцилятора без затуханий и без действия внешней силы

```python
w=np.sqrt(5.1)
g=0

def f(t):
    f= sin(0.00*t)
    return f

def dx(x,t):
    dx1=x[1]
    dx2= -w*w*x[0]-g*x[1]-f(t)
    return [dx1, dx2]
```

```python
t0=0
t= np.arange(t0, 38, 0.05)
x0=[0.9,-1.9]
```

```python
x=odeint(dx, x0, t)
y=[[elem[0] for elem in x],[elem[1] for elem in x]]
```

```python
pltl.grid()
pltl.title('Решение для колебаний гармонического осцилятора без затуханий и без действия внешней силы')
pltl.plot(t,x)
```

```python
pltl.grid()
pltl.title('Фазовый портрет для колебаний гармонического осцилятора без затуханий и без действия внешней силы')
pltl.ylabel('y')
pltl.xlabel('x')
pltl.plot(y[0],y[1])
```
