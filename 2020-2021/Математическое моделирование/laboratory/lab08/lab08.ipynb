```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math
```

```python
p_cr=33 #
tau1=27 #
p1=11.3
tau2=24
p2=12.5
V=83
q=1

t0=0
tmax=15
t=np.arange(t0, tmax, 0.01)
x0=[8.5,9.1]
```

```python
a1=p_cr/(tau1*tau1*p1*p1*V*q)
a2=p_cr/(tau2*tau2*p2*p2*V*q)
b=p_cr/(tau1*tau1*p1*p1*tau2*tau2*p2*p2*V*q)
c1=(p_cr-p1)/(tau1*p1)
c2=(p_cr-p2)/(tau2*p2)

def dx1(x,t):
    dx1=x[0]-(a1/c1)*x[0]*x[0]-(b/c1)*x[0]*x[1]
    dx2=(c2/c1)*x[1]-(a2/c1)*x[1]*x[1]-(b/c1)*x[0]*x[1]
    return [dx1, dx2]

x=odeint(dx1, x0, t)
```

```python
plt.plot(t,x)
plt.legend('12')
```

```python
def dx2(x,t):
    dx1=x[0]-(a1/c1)*x[0]*x[0]-(b/c1+0.00019)*x[0]*x[1]
    dx2=(c2/c1)*x[1]-(a2/c1)*x[1]*x[1]-(b/c1)*x[0]*x[1]
    return [dx1, dx2]

x=odeint(dx2, x0, t)
```

```python
plt.plot(t,x)
plt.legend('12')
```
