```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
N = 11600
t0 = 0
tmax = 200
I0 = 260
R0 = 48
S0 = N - I0 - R0

alpha = 0.04
beta = 0.07

x0 = np.array([S0, I0, R0])
t = np.arange(t0, tmax, 0.01)

def dx_less(x, t):
    dx1 = 0.0
    dx2 = -beta*x[1]
    dx3= beta*x[1]
    return [dx1, dx2, dx3]

def dx_greater(x, t):
    dx1 = -alpha*x[0]
    dx2 = alpha*x[0] - beta*x[1]
    dx3= beta*x[1]
    return [dx1, dx2, dx3]

y_less = odeint(dx_less, x0, t)
y_greater = odeint(dx_greater, x0, t)
```

```python
plt.plot(t, y_less[:, 0])
plt.plot(t, y_less[:, 1])
plt.plot(t, y_less[:, 2])
plt.title("I(0)<=I*")
plt.ylabel("Численность людей")
plt.xlabel("Время")
plt.legend(["S(t)", "I(t)", "R(t)"])
```

```python
plt.plot(t, y_greater[:, 0])
plt.plot(t, y_greater[:, 1])
plt.plot(t, y_greater[:, 2])
plt.title("I(0)>I*")
plt.ylabel("Численность людей")
plt.xlabel("Время")
plt.legend(["S(t)", "I(t)", "R(t)"])
```
